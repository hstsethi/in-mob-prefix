## Importing Neccessary Librariries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Plotting Mobile Number With Prefixes 9xxx By Circle

```python
nine_data = pd.read_csv("9xxx-in-mob-prefix.csv")
nine_data_circle = nine_data["Circle"].value_counts().plot(kind="bar", title="Mobile Numbers By Prefix 9xxx By Circle")
nine_data_operator = nine_data["Operator"].value_counts().plot(kind="bar", title="Operators With Mobile Numbers Starting From Prefix 9xxx")
```
